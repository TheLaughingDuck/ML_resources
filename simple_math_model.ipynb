# Training a model to fit a mathematical function

```python
###### Create target function ######
def target_func(x):
    return x**2

###### Generate training data ######
import numpy as np
import random

n_train_samples = 2000
X_train = np.array([[1000*random.random()] for i in range(n_train_samples)])
y_train = np.array([target_func(value) for value in X_train[:,0].tolist()])

###### Generate test data ######
X_test = np.array([[i] for i in range(1000)])
y_test = np.array([target_func(value) for value in X_test[:,0].tolist()])
```

```python
###### Create and train a model ######
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(max_iter=2000, hidden_layer_sizes=[100, 80, 50, 40, 10, 3], activation="relu")
model.fit(X_train, y_train)
```

```python
###### Predict test data ######
predictions = model.predict(X_test)

###### Plot predictions ######
from matplotlib import pyplot as plt

fig, axs = plt.subplots(nrows=1)
axs.plot(X_test[:,0].tolist(), y_test.tolist())
axs.plot(X_test[:,0].tolist(), predictions, color="red")
```
